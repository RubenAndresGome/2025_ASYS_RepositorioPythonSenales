```python
# %% [markdown]
# # Transformada de Fourier de una Seal Exponencial
# 
# En este notebook, calculamos y visualizamos la transformada de Fourier de la funcin:
# 
# $$
# f(t) = 
# \begin{cases} 
# e^{-\alpha t}, & t > 0 \\
# 0, & t < 0 
# \end{cases}, \quad \alpha > 0
# $$
# 
# Compararemos el resultado numrico (usando FFT) con la solucin analtica:
# 
# $$
# F(\omega) = \frac{1}{\alpha + i\omega}
# $$

# %%
# Importar libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from thinkdsp import Wave, Spectrum
import warnings
warnings.filterwarnings('ignore')  # Para limpiar la salida

# Configuracin general de las grficas
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# %% [markdown]
# ## 1. Definicin de la seal en el dominio del tiempo

# %%
# Parmetros de la simulacin
alpha = 1.0          # Parmetro de decaimiento
duration = 10.0      # Duracin total de la seal (en segundos)
framerate = 1000     # Muestreo (puntos por segundo)
dt = 1 / framerate   # Intervalo de muestreo

# Crear el eje de tiempo (desde 0 hasta duration)
t = np.linspace(0, duration, int(duration * framerate), endpoint=False)

# Generar la seal: f(t) = e^{-t} para t  0
ys = np.exp(-alpha * t)

# Crear objeto Wave de ThinkDSP
wave = Wave(ys, framerate=framerate)

# Visualizar la seal en el dominio del tiempo
plt.figure(figsize=(10, 4))
wave.plot()
plt.title(f'Seal en el dominio del tiempo: $f(t) = e^{{-{alpha}t}}$ para $t \\geq 0$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# %% [markdown]
# ## 2. Clculo de la transformada de Fourier

# %%
# Calcular la transformada de Fourier completa (incluyendo frecuencias negativas)
spectrum = wave.make_spectrum(full=True)

# Aplicar fftshift para ordenar las frecuencias de -f_max a +f_max
n = len(spectrum.hs)
fs_shifted = np.fft.fftshift(spectrum.fs)
hs_shifted = np.fft.fftshift(spectrum.hs)

# Escalar por dt para aproximar la integral continua (FT  sum(ys) * dt)
hs_shifted_scaled = hs_shifted * dt

# Calcular la solucin analtica F() = 1/( + i)
# Nota:  = 2f (frecuencia angular en rad/s)
f_analytical = fs_shifted
omega = 2 * np.pi * f_analytical
F_analytical = 1 / (alpha + 1j * omega)

# Magnitud y fase analticas
mag_analytical = np.abs(F_analytical)
phase_analytical = np.angle(F_analytical)

# %% [markdown]
# ## 3. Visualizacin de resultados

# %%
# Crear figura con 3 subplots
plt.figure(figsize=(14, 10))

# 1. Dominio del tiempo (ya calculado anteriormente)
plt.subplot(3, 1, 1)
wave.plot()
plt.title(f'Seal en el dominio del tiempo: $f(t) = e^{{-{alpha}t}}$ para $t \\geq 0$')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True, linestyle='--', alpha=0.7)

# 2. Magnitud del espectro
plt.subplot(3, 1, 2)
plt.plot(fs_shifted, np.abs(hs_shifted_scaled), 
         label='Numrico (FFT escalado)', linewidth=2)
plt.plot(f_analytical, mag_analytical, 'r--', 
         label='Analtico: $|F(\\omega)| = \\frac{1}{\\sqrt{\\alpha^2 + \\omega^2}}$', 
         linewidth=2.5)
plt.xlim(-10, 10)  # Enfocar en frecuencias bajas donde hay ms energa
plt.title('Magnitud del espectro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# 3. Fase del espectro
plt.subplot(3, 1, 3)
plt.plot(fs_shifted, np.angle(hs_shifted_scaled), 
         label='Numrico (FFT escalado)', linewidth=2)
plt.plot(f_analytical, phase_analytical, 'r--', 
         label='Analtico: $\\phi(\\omega) = -\\arctan(\\omega/\\alpha)$', 
         linewidth=2.5)
plt.xlim(-10, 10)
plt.ylim(-np.pi/2 - 0.1, np.pi/2 + 0.1)
plt.yticks([-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2],
           [r'$-\pi/2$', r'$-\pi/4$', '0', r'$\pi/4$', r'$\pi/2$'])
plt.title('Fase del espectro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 4. Anlisis detallado de la comparacin

# %%
# Crear una comparacin detallada en frecuencias bajas
plt.figure(figsize=(14, 5))

# Magnitud - comparacin detallada
plt.subplot(1, 2, 1)
plt.plot(fs_shifted, np.abs(hs_shifted_scaled), 'b-', linewidth=1.5)
plt.plot(f_analytical, mag_analytical, 'r--', linewidth=2)
plt.xlim(-2, 2)
plt.title('Magnitud del espectro (detalle)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(['Numrico', 'Analtico'])

# Calcular error relativo en magnitud
error_mag = np.abs(np.abs(hs_shifted_scaled) - mag_analytical) / mag_analytical
error_mag = np.where(mag_analytical > 1e-5, error_mag, np.nan)  # Evitar divisin por cero

# Error relativo
plt.subplot(1, 2, 2)
plt.semilogy(fs_shifted, error_mag, 'g-', linewidth=2)
plt.xlim(-2, 2)
plt.ylim(1e-10, 1)
plt.title('Error relativo en magnitud')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Error relativo')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 5. Verificacin del valor en DC (f = 0)

# %%
# Verificar el valor en frecuencia cero (DC)
dc_numerical = np.abs(hs_shifted_scaled[np.isclose(fs_shifted, 0)])[0]
dc_analytical = mag_analytical[np.isclose(f_analytical, 0)][0]

print(f"Valor en DC (f = 0):")
print(f" - Numrico (FFT): {dc_numerical:.6f}")
print(f" - Analtico: {dc_analytical:.6f}")
print(f" - Error relativo: {np.abs(dc_numerical - dc_analytical)/dc_analytical:.2e}")

# Verificar la energa total usando el teorema de Parseval
energy_time = np.sum(wave.ys**2) * dt
energy_freq = np.sum(np.abs(hs_shifted_scaled)**2) * (fs_shifted[1] - fs_shifted[0]) * 2  # *2 por frecuencias negativas

print("\nTeorema de Parseval (conservacin de energa):")
print(f" - Energa en dominio del tiempo: {energy_time:.6f}")
print(f" - Energa en dominio de la frecuencia: {energy_freq:.6f}")
print(f" - Error relativo: {np.abs(energy_time - energy_freq)/energy_time:.2e}")

# %% [markdown]
# ## 6. Efecto del parmetro alpha

# %%
# Visualizar cmo cambia el espectro con diferentes valores de alpha
alphas = [0.5, 1.0, 2.0]
plt.figure(figsize=(12, 5))

# Magnitud para diferentes alpha
plt.subplot(1, 2, 1)
for alpha_val in alphas:
    F_val = 1 / (alpha_val + 1j * omega)
    plt.plot(f_analytical, np.abs(F_val), 
             label=f' = {alpha_val}', linewidth=2)

plt.xlim(-10, 10)
plt.title('Efecto de  en la magnitud del espectro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Fase para diferentes alpha
plt.subplot(1, 2, 2)
for alpha_val in alphas:
    F_val = 1 / (alpha_val + 1j * omega)
    plt.plot(f_analytical, np.angle(F_val), 
             label=f' = {alpha_val}', linewidth=2)

plt.xlim(-10, 10)
plt.ylim(-np.pi/2 - 0.1, np.pi/2 + 0.1)
plt.yticks([-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2],
           [r'$-\pi/2$', r'$-\pi/4$', '0', r'$\pi/4$', r'$\pi/2$'])
plt.title('Efecto de  en la fase del espectro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## Conclusin
# 
# Este notebook demuestra cmo:
# 
# 1. La FFT (con escalamiento apropiado) puede aproximar con alta precisin la transformada de Fourier continua
# 2. La solucin analtica $ F(\omega) = \\frac{1}{\\alpha + i\\omega} $ coincide casi perfectamente con el resultado numrico
# 3. El parmetro $\\alpha$ controla el ancho de banda de la seal:
#    - Valores pequeos de $\\alpha$  decaimiento lento en tiempo  espectro angosto en frecuencia
#    - Valores grandes de $\\alpha$  decaimiento rpido en tiempo  espectro ancho en frecuencia
# 
# **Aplicaciones prcticas:** Este tipo de anlisis es fundamental en:
# - Diseo de filtros analgicos y digitales
# - Anlisis de circuitos RC y sistemas de primer orden
# - Procesamiento de seales en telecomunicaciones
# - Estudio de sistemas de control
```
